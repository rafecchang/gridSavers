```python
import pandas as pd 
import numpy as np 
import altair as alt
alt.data_transformers.enable("vegafusion")
```

```python
df = pd.read_csv('data/dataverse_files/Residential_1.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['datetime'] = df['date'] + pd.to_timedelta(df['hour'] - 1, unit='h')
```

```python
df_eda = alt.Chart(df).mark_line(color="#333").encode(
    x=alt.X("datetime:T", title="hour"),  
    y=alt.Y("energy_kWh:Q", title="Energy (kWh)")  
)

df_eda
```

```python
df06010212 = df[df['date'].isin(['2012-06-01', '2012-06-02'])]
# df06010212
```

```python
df06010212_eda = alt.Chart(df06010212).mark_line(color="#333").encode(
    x=alt.X("datetime:T", title="hour"),  
    y=alt.Y("energy_kWh:Q", title="Energy (kWh)")  
)
df06010212_eda
```

BC Hydro Time of Day Pricing (https://app.bchydro.com/accounts-billing/rates-energy-use/electricity-rates/residential-rates/time-of-day.html):
- 16:00 - 21:00 + 5 cent per kWh (peak)
- 21:00 - 23:00 + 0 cent per kWh 
- 23:00 - 07:00 - 5 cent per kWh (off-peak)
- 07:00 - 16:00 + 0 cent per kWh 

# Simple Simulation 1 
Assume:
- Home owner has a Tesla Model 3 Standard Range Plus of 50 kWh 
- The charging period over off peak is sufficient to supply all capacity needed 
- The energy output can match with energy demand 



```python
saving_06010212 = df06010212[(df06010212['hour'] >= 16) & (df06010212['hour'] <= 20)]['energy_kWh'].sum()
saving_06010212
```

On 2012-06-01, and 2012-06-02, this household used 9.077 kWh of electricity in between 16:00 to 21:00. If the household is participated in BC Hydro's Time of Day Pricing and has a battery system, they can save up to 9.077 dollars.
